## Descriptive Statistics

>Descriptive statistics will help us to explore simplified interpretation of our data

Explore the total number of passengers onboard the titanic from the various embarkation centers ( Southanpton, Cherbourg and Queensland

```python
embarkation_counts = titanic["embarked"].value_counts()
```

```python
embarkation_counts
```

### Create a bar chart to show the above visually and also a piechart to show each contribution to the whole

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(8,6))
titanic["embarked"].value_counts().plot(kind="bar")
plt.title("Passenger by Embarkation Center")
plt.xlabel("Embarkation Center")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
plt.figure(figsize=(8,6))
titanic["embarked"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Passenger by Embarkation Center")
plt.show()
```

### Explore embarkation by gender and by passenger class  (**`Cross tabulation`**)

### **`Embarkation by Gender`**

```python
gender_embarkation = pd.crosstab(titanic["sex"],titanic["embarked"])
```

```python
gender_embarkation
```

### **`Embarkation by passenger class`**

```python
pclass_embarkation = pd.crosstab(titanic["passenger_class"], titanic["embarked"])
```

```python
pclass_embarkation
```

### **`Sex by Passenger class`**

```python
sex_pclass = pd.crosstab(titanic["sex"], titanic["passenger_class"])
```

```python
sex_pclass
```

### **`Sex by fate (Survival status)`**

```python
sex_survival = pd.crosstab(titanic["sex"], titanic["survived"])
```

```python
sex_survival
```
